```python
import spacy
from spacy.training import Example

# Creamos un modelo vacío
nlp = spacy.blank("xx")

# Añadimos detector de entidades
ner = nlp.add_pipe("ner")

# Añadimos etiquetas
ner.add_label("PERSON")
ner.add_label("LOC")
ner.add_label("ORG")
```

```python
TRAIN_DATA = [
    ("Yusuf yezga deg Rabat", {"entities": [(0, 5, "PERSON"), (16, 21, "LOC")]}),
    ("Fatima teddut ɣer Paris", {"entities": [(0, 6, "PERSON"), (18, 23, "LOC")]}),
    ("Mohamed yezga deg Madrid", {"entities": [(0, 7, "PERSON"), (16, 22, "LOC")]}),
]
```

```python
import random

optimizer = nlp.begin_training()

for i in range(100):  # más vueltas
    random.shuffle(TRAIN_DATA)
    for text, annotations in TRAIN_DATA:
        example = Example.from_dict(nlp.make_doc(text), annotations)
        nlp.update([example], sgd=optimizer)
```

```python
doc = nlp("Yusuf yedda ɣer Paris")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
